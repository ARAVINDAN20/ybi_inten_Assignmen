```python
# step 1:  import library
import pandas as pd
```

```python
# step 2 : import dataset
salary = pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Salary%20Data.csv")
```

```python
salary.head()
```

```python
salary.info()
```

```python
# step 3: define [your problem] target(y) and features(x)
```

```python
salary.columns
```

```python
y = salary ["Salary"]
x= salary[["Experience Years"]]
```

```python
# step 4 : spliting the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 2529)
```

```python
# step 5 : select model - is called or we use -linear regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# step 6: traine model (fit model)
model.fit(x_train,y_train)
```

```python
# Intercept of this model       {y = c+ m x}  Predictive model
model.intercept_
```

```python
# slope of this model
model.coef_
```

```python
# step 7 : prediction     salary = 27010.9 + 9343.4 * [first value i.e year] 6.8
y_pred = model.predict(x_test)    #salary =  90545.86017355
```

```python
x_test
```

```python
# step 8 : accuracy
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
# 6% is th error so our  ML model is 94% accuracy
```
